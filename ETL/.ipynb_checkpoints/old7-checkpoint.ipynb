```python
### Imports necesarios
import pyodbc #odbc para sqlserver
import time #librería necesaria para medir el tiempo de ejecución 
import datetime
import pandas as pd
import numpy as np
from functools import reduce

# Carga de datos en MySQL
from sqlalchemy import create_engine
import pymysql
import os

user = os.environ['chcprocuser']
passw = os.environ['chcprocpass']
server = os.environ['chcprocserver']
db = os.environ['chcprocdb']

# Inicio del "cronómetro"
tInicio = time.time()

# Apertura la conexión a SQL SERVER
cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER='+server+';DATABASE='+db+';UID='+user+';PWD='+passw)

# Apertura la conexión a MYSQL
engine = create_engine('mysql+pymysql://server:server@192.168.2.2:3306/espaciopublico', echo = False)

cursor = cnxn.cursor()
```

```python
sqldf = '''
SELECT porCreationDate
,porSendDate
,porJustifyType
,porBuyerEnterprise
,porSellerEnterprise
,porSourceDocumentNumber
,item.poiCurrency
,item.poiTotalAmount
,eroc.exrValue 'cambioOC'
,erit.exrValue 'cambioItem'
FROM DCCPProcurement.dbo.prcPOHeader 
INNER JOIN DCCPProcurement.dbo.prcPOItem item ON porID=item.poiOrder
LEFT JOIN DCCPPlatform.dbo.gblExchangeRate eroc ON DCCPProcurement.dbo.prcPOHeader.porCurrency=eroc.exrCurrency AND (DCCPProcurement.dbo.prcPOHeader.porSendDate BETWEEN eroc.exrDateFrom AND eroc.exrDateTo)
LEFT JOIN DCCPPlatform.dbo.gblExchangeRate erit ON item.poiCurrency=erit.exrCurrency AND (DCCPProcurement.dbo.prcPOHeader.porSendDate BETWEEN erit.exrDateFrom AND erit.exrDateTo)
WHERE year(DCCPProcurement.dbo.prcPOHeader.porSendDate) in (2019, 2020)
AND porBuyerStatus NOT IN (1,2,3,8,9,10,11,14,15)
'''

df = pd.read_sql(sqldf, cnxn)
```

```python
df[df['porJustifyType'] == 703]
```

```python
# TIPO DE CONTRATACION
df['Tipodeproceso'] = 'Trato Directo'
df.loc[df['porJustifyType'] == 703, 'Tipodeproceso'] = 'Convenio Marco'
df.loc[df['porJustifyType'] == 702, 'Tipodeproceso'] = 'Licitacion Privada'
df.loc[df['porJustifyType'] == 701, 'Tipodeproceso'] = 'Licitacion Publica'
df.loc[df['porJustifyType'] == 1301, 'Tipodeproceso'] = '.'
df.loc[df['porJustifyType'] == 1401, 'Tipodeproceso'] = '.'
df.loc[df['porSourceDocumentNumber'].isna(), 'porSourceDocumentNumber'] = 0
df.head(2)
```

```python
df[df['Tipodeproceso'] == 'Convenio Marco']
```

```python
# MONTOS: (ITEM)
df['MontototalenCLP'] = 10000000000.0
df.loc[df['poiCurrency'] != 'CLP', 'MontototalenCLP'] = df['poiTotalAmount'] * df['cambioItem']
df.loc[df['poiCurrency'] == 'CLP', 'MontototalenCLP'] = df['poiTotalAmount']
```

```python
df['year'] = pd.DatetimeIndex(df['porSendDate']).year
df['MontototalenCLP'] = round(df['MontototalenCLP'])

df.head()
```

```python
dfN_Empresas_Ganadoras = df[['porSellerEnterprise', 'porBuyerEnterprise', 'year']]
dfN_Empresas_Ganadoras = dfN_Empresas_Ganadoras.drop_duplicates()
dfN_Empresas_Ganadoras = dfN_Empresas_Ganadoras.groupby(['porBuyerEnterprise', 'year'])['porSellerEnterprise'].count().reset_index()
dfN_Empresas_Ganadoras
```

```python
dfNyMonto_Contratos_TC = df.copy()
dfNyMonto_Contratos_TC = dfNyMonto_Contratos_TC.groupby(['porBuyerEnterprise', 'Tipodeproceso', 'porSourceDocumentNumber', 'porSellerEnterprise', 'year'])['MontototalenCLP'].agg({'count', 'sum'}).reset_index()
dfNyMonto_Contratos_TC = dfNyMonto_Contratos_TC.rename(columns = {'count': 'Cantidad', 'sum': 'Monto'})
dfNyMonto_Contratos_TC.loc[dfNyMonto_Contratos_TC['Tipodeproceso'] == 'Licitacion Publica', 'Cantidad'] = 1
dfNyMonto_Contratos_TC.loc[dfNyMonto_Contratos_TC['Tipodeproceso'] == 'Licitacion Privada', 'Cantidad'] = 1
```

```python
dfNyMonto_Contratos_TC[dfNyMonto_Contratos_TC['Tipodeproceso']=='Convenio Marco']
```

```python
dfNyMonto_Contratos_TC_year = dfNyMonto_Contratos_TC.groupby(['porBuyerEnterprise', 'Tipodeproceso', 'year'])['Cantidad', 'Monto'].sum().reset_index()
dfNyMonto_Contratos_TC_year
```

```python
dfNyMonto_Contratos_Proveedor = dfNyMonto_Contratos_TC.groupby(['porBuyerEnterprise', 'porSellerEnterprise', 'year'])['Cantidad', 'Monto'].sum().reset_index()
dfNyMonto_Contratos_Proveedor
```

```python
dfN_Contratos_Year = dfNyMonto_Contratos_TC.groupby(['porBuyerEnterprise', 'year'])['Cantidad'].sum().reset_index()
dfN_Contratos_Year
```

## 4) Número de empresas ganadoras diferentes por cada 100 contratos.

```python
## 4) Número de empresas ganadoras diferentes por cada 100 contratos. 

aux1 = dfN_Contratos_Year.copy()
aux2 = dfN_Empresas_Ganadoras.copy()

ind4 = pd.merge(aux1, aux2, how= 'left', left_on=["porBuyerEnterprise", "year"], right_on=["porBuyerEnterprise", "year"])
ind4['Ratio_Ganadores_Contratos']= round(100*ind4['porSellerEnterprise']/ind4['Cantidad'])
ind4 = ind4[['porBuyerEnterprise', 'year', 'Ratio_Ganadores_Contratos']]
ind4['Ratio_Ganadores_Contratos'] = ind4['Ratio_Ganadores_Contratos'].replace(np.inf, 0)

# guardamos en la tabla
ind4.to_sql(con=engine, name='indicador4', if_exists='replace', index=False)
ind4.sort_values('Ratio_Ganadores_Contratos')
```

## 5) Porcentaje de procedimientos que utilizaron adjudicación directa o INV3. 

```python
ind5 = dfNyMonto_Contratos_TC_year.copy()
ind5G = ind5.groupby(by=['porBuyerEnterprise', 'year'])['Cantidad'].sum().reset_index()
ind5['CantDirecta'] = 0
ind5.loc[ind5['Tipodeproceso'].isin(['Licitacion Privada', 'Trato Directo']), 'CantDirecta'] = ind5['Cantidad']
ind5H = ind5.groupby(by=['porBuyerEnterprise', 'year'])['CantDirecta'].sum().reset_index()
ind5F = pd.merge(ind5G, ind5H, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind5F['PorcProcDirecto'] = 100 * ind5F['CantDirecta'] / ind5F['Cantidad']
ind5F = ind5F.drop(columns=['Cantidad', 'CantDirecta'])

# guardamos en la tabla
ind5F.to_sql(con=engine, name='indicador5', if_exists='replace', index=False)                            

ind5F[ind5F['porBuyerEnterprise'] == '6957']
```

## 6) Porcentaje del monto adjudicado por adjudicación directa o INV3.

```python
ind6 = dfNyMonto_Contratos_TC_year.copy()
ind6G = ind6.groupby(by=['porBuyerEnterprise', 'year'])['Monto'].sum().reset_index()
ind6['MontoDirecta'] = 0
ind6.loc[ind6['Tipodeproceso'].isin(['Licitacion Privada', 'Trato Directo']), 'MontoDirecta'] = ind6['Monto']
ind6H = ind6.groupby(by=['porBuyerEnterprise', 'year'])['MontoDirecta'].sum().reset_index()
ind6F = pd.merge(ind6G, ind6H, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind6F['PorcMontoDirecto'] = 100 * ind6F['MontoDirecta'] / ind6F['Monto']
ind6F = ind6F.drop(columns=['Monto', 'MontoDirecta'])

# guardamos en la tabla
ind6F.to_sql(con=engine, name='indicador6', if_exists='replace', index=False)
ind6F[ind6F['porBuyerEnterprise'] == '6957']
```

## 7) Cambio en la tendencia del porcentaje de adjudicaciones directas (No se hace porque esta contenido en 6).
## 8) IHH por número de contratos. 

```python
ind8 = dfNyMonto_Contratos_Proveedor.copy()
ind8A = ind8.groupby(by=['porBuyerEnterprise', 'year'])['Cantidad'].sum().reset_index()
ind8B = pd.merge(ind8, ind8A, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind8B['indHHC'] = pow(round(ind8B['Cantidad_x'] / ind8B['Cantidad_y'], 2), 2)
ind8F = ind8B.groupby(by=['porBuyerEnterprise', 'year'])['indHHC'].sum().reset_index()

# guardamos en la tabla
ind8F.to_sql(con=engine, name='indicador8', if_exists='replace', index=False)
ind8F[ind8F['porBuyerEnterprise'] == '6957']
```

## 9) IHH por monto de contratos. 

```python
ind9 = dfNyMonto_Contratos_Proveedor.copy()
ind9A = ind9.groupby(by=['porBuyerEnterprise', 'year'])['Monto'].sum().reset_index()
ind9B = pd.merge(ind8, ind9A, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind9B['indHHM'] = pow(round(ind9B['Monto_x'] / ind9B['Monto_y'] * 100, 2), 2)
ind9F = ind9B.groupby(by=['porBuyerEnterprise', 'year'])['indHHM'].sum().reset_index()

# guardamos en la tabla
ind9F.to_sql(con=engine, name='indicador9', if_exists='replace', index=False)

ind9F[ind9F['porBuyerEnterprise'] == '6957']
```

## 10) ID por número de contratos. 

```python
ind10 = dfNyMonto_Contratos_Proveedor.copy()
ind10A = pd.merge(ind10, ind8B, how='left', left_on=['porBuyerEnterprise', 'year', 'porSellerEnterprise'], right_on=['porBuyerEnterprise', 'year', 'porSellerEnterprise'])
ind10B = pd.merge(ind10A, ind8F, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind10B['ID_Cantidad'] = pow((ind10B['indHHC_x']/ind10B['indHHC_y']), 2) * 100
ind10F = ind10B.groupby(by=['porBuyerEnterprise', 'year'])['ID_Cantidad'].mean().reset_index()

# guardamos en la tabla
ind10F.to_sql(con=engine, name='indicador10', if_exists='replace', index=False)

ind10F[ind10F['porBuyerEnterprise'] == '6957']
```

## 11) ID por monto de contratos.

```python
ind11 = dfNyMonto_Contratos_Proveedor.copy()
ind11A = pd.merge(ind11, ind9B, how='left', left_on=['porBuyerEnterprise', 'year', 'porSellerEnterprise'], right_on=['porBuyerEnterprise', 'year', 'porSellerEnterprise'])
ind11B = pd.merge(ind11A, ind9F, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind11B['ID_Monto'] = pow((ind11B['indHHM_x']/ind11B['indHHM_y']), 2) * 100
ind11F = ind11B.groupby(by=['porBuyerEnterprise', 'year'])['ID_Monto'].mean().reset_index()
# guardamos en la tabla
ind11F.to_sql(con=engine, name='indicador11', if_exists='replace', index=False)
ind11F[ind11F['porBuyerEnterprise'] == '6957']
```

## 14) Índice de concentración de las 4 empresas con más procedimientos ganados.

```python
ind14 = dfNyMonto_Contratos_Proveedor.copy()
ind14A = ind14.groupby(by=['porBuyerEnterprise', 'year'])['Cantidad'].sum().reset_index()
ind14B = ind14.merge(ind14A, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind14B['Part'] = ind14B['Cantidad_x'] / ind14B['Cantidad_y'] * 100
ind142019 = ind14B[ind14B['year'] == 2019]
ind142020 = ind14B[ind14B['year'] == 2020]
ind142019['Rank'] = ind142019.groupby('porBuyerEnterprise')['Part'].rank(ascending=False, method='dense')
ind142020['Rank'] = ind142020.groupby('porBuyerEnterprise')['Part'].rank(ascending=False, method='dense')
ind14B['Rank'] = ind14B.groupby('porBuyerEnterprise')['Part'].rank(ascending=False, method='dense')
ind142019C = ind142019[ind142019['Rank']<=4]
ind142020C = ind142020[ind142020['Rank']<=4]
ind14C = pd.concat([ind142019C,ind142020C])
ind14F = ind14C.groupby(by=['porBuyerEnterprise', 'year'])['Part'].sum().reset_index()

# guardamos en la tabla
ind14F.to_sql(con=engine, name='indicador14', if_exists='replace', index=False)
ind14F[ind14F['porBuyerEnterprise'] == '6957']
```

## 15) Índice de concentración de las 4 empresas con más monto ganados. 

```python
ind15 = dfNyMonto_Contratos_Proveedor.copy()
ind15A = ind15.groupby(by=['porBuyerEnterprise', 'year'])['Monto'].sum().reset_index()
ind15B = ind15.merge(ind15A, how='left', left_on=['porBuyerEnterprise', 'year'], right_on=['porBuyerEnterprise', 'year'])
ind15B['Part'] = ind15B['Monto_x'] / ind15B['Monto_y'] * 100
ind152019 = ind15B[ind15B['year'] == 2019]
ind152020 = ind15B[ind15B['year'] == 2020]
ind152019['Rank'] = ind152019.groupby('porBuyerEnterprise')['Part'].rank(ascending=False, method='dense')
ind152020['Rank'] = ind152020.groupby('porBuyerEnterprise')['Part'].rank(ascending=False, method='dense')
ind152019C = ind152019[ind152019['Rank']<=4]
ind152020C = ind152020[ind152020['Rank']<=4]
ind15C = pd.concat([ind152019C,ind152020C])
ind15F = ind15C.groupby(by=['porBuyerEnterprise', 'year'])['Part'].sum().reset_index()

# guardamos en la tabla
ind15F.to_sql(con=engine, name='indicador15', if_exists='replace', index=False)
ind15F[ind15F['porBuyerEnterprise'] == '6957']
```
