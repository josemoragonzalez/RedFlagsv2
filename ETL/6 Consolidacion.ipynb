```python
# importamos las librerias necesarias
import mysql.connector
import pandas as pd
import time #librería necesaria para medir el tiempo de ejecución 
import datetime
from joblib import dump, load # importamos la libreria correspodiente para poder cargar el modelo elaborado a partir de datos históricos

# Inicio del "cronómetro"
tInicio = time.time()

# generamos el conector
cnxn = mysql.connector.connect(host='192.168.2.2', database='nuevos', user='server', password='server')

# query para traer las distintas tablas en que se almcenaron los resultados de los indicadores
sqlGarantias = '''select * from requisitoGarantía'''
sqlRenovaContrato = '''select * from renovacionContrato'''
sqlExperienciaExcesiva = '''select * from requisitoExperiencia'''
sqlLimites = '''select * from montoCercanoUmbral'''


garantias = pd.read_sql(sqlGarantias, cnxn)
renovaContrato = pd.read_sql(sqlRenovaContrato, cnxn)
experienciaExcesiva = pd.read_sql(sqlExperienciaExcesiva, cnxn)
limites = pd.read_sql(sqlLimites, cnxn)

# hacemos un merge para unir todos los conjuntos
dfs = [renovaContrato, experienciaExcesiva, limites]

inicialElab = garantias.copy()
for i in dfs:
    inicialElab = inicialElab.merge(i, how= 'left')

# Carga de datos en MySQL
from sqlalchemy import create_engine
import pymysql

engine = create_engine('mysql+pymysql://server:server@192.168.2.2:3306/nuevos', echo = False)
inicialElab.to_sql(con=engine, name='Elaboracion', if_exists='replace', index=False)

# Término del "crónometro" y transformaciones a hora, minutos y segundos
tFinal = time.time()
tSegundosGenerico = tFinal - tInicio
tFormateado = str(datetime.timedelta(seconds=tSegundosGenerico))

# Resultados de la primera extracción
print("Terminado, se procesaron "+ str(len(inicialElab.index)) + " licitaciones, en "+ tFormateado)
```
