## Elaboración
Pasos para realizar el clustering a las licitaciones en el proceso de Elaboración

### Importación de datos y generación del dataset unificado
En el siguiente código se leen cada una de las tablas que contienen los valores de indicadores pertenecientes a la etapa de Elaboración, para luego dejarlas en un dataset inicialElab

```python
# importamos las librerias necesarias
import mysql.connector
import pandas as pd

# generamos el conector
cnxn = mysql.connector.connect(host='192.168.2.2', database='redflags', user='server', password='server')

# query para traer las distintas tablas en que se almcenaron los resultados de los indicadores
sqlGarantias = '''select * from garantias'''
sqlRenovaContrato = '''select * from renovaContrato'''
sqlCriterioIncompleto = '''select * from qCriterio'''
sqlExperienciaExcesiva = '''select * from experienciaExcesiva'''
sqlNoExtranjero = '''select * from noExtranjero'''
sqlReqGeografico = '''select * from reqGeografico'''
sqlLimites = '''select * from indValorUTM'''


garantias = pd.read_sql(sqlGarantias, cnxn)
renovaContrato = pd.read_sql(sqlRenovaContrato, cnxn)
criterioIncompleto = pd.read_sql(sqlCriterioIncompleto, cnxn)
experienciaExcesiva = pd.read_sql(sqlExperienciaExcesiva, cnxn)
noExtranjero = pd.read_sql(sqlNoExtranjero, cnxn)
reqGeografico = pd.read_sql(sqlReqGeografico, cnxn)
limites = pd.read_sql(sqlLimites, cnxn)

# hacemos un merge para unir todos los conjuntos
dfs = [renovaContrato, criterioIncompleto, experienciaExcesiva, noExtranjero, reqGeografico, limites]

inicialElab = garantias.copy()
for i in dfs:
    inicialElab = inicialElab.merge(i, how= 'left')
```

### Estadísticas y análisis exploratorio del dataset
El primer análisis corresponde a una descripción con estadísticos estándares como el promedio, la desviación estándar, minimos y máximos, junto con los cuartiles

```python
inicialElab.describe()
```

Se puede apreciar que de acuerdo al promedio el indicador con menos positivos es el de Requisito de Nacionalidad mean = 2.923202e-05 y el que más positivos tiene es el de Garantías mean = 7.448784e-01

Posterioremente tenemos una tabla pivote que permite visualzar los escenarios en que se intersectan los indicadores de riesgo para las distintas licitaciones

```python
tabla = pd.pivot_table(inicialElab, values='rbhCode', index=['indGarantia', 'indCriterio', 'indExperiencia', 'RequisitoNacionalidad', 'RequisitoGeografico', 'indvalorEnUTM'],
                    columns=['indRenovacion'], aggfunc='count')
tabla
```

Como se puede apreciar en la tabla anterior, no existen licitaciones que tengan positivos todos los indicadores y existen 240.299 licitaciones que no presentan ningun tipo de warning

### Clustering, ¿Cuantos?
En el siguiente código se utilizará la librería sci-kit learn con  el algoritmo k-.means para agrupar a las licitaciones de acuerdo a sus valores en cada indicador, asimismo, utilizaramos la regla del codo para estimnar la cantidad de clusters 

```python
#importamos las librerias necesarias
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(inicialElab[['indGarantia','indRenovacion','indCriterio','indExperiencia','RequisitoNacionalidad','RequisitoGeografico','indvalorEnUTM']])
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
```

Vemos en el gráfico que a partir 4 o 5 grupos la suma de los cuadrados  de las distancias de los puntos a sus centroides se estabiliza, lo que hace que un número mayor de grupos agregue muy poco valor.

Por ello ahora calculamos con 5 clusters:

```python
kmeansElabora = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeansElabora.fit(inicialElab[['indGarantia','indRenovacion','indCriterio','indExperiencia','RequisitoNacionalidad','RequisitoGeografico','indvalorEnUTM']])
```

Contamos la cantidad de licitaciones por cluster y vemos que no hay clusters con un número irracional de elementos, porm lo que, seguimos adelante con el proceso

```python
#Conteo de cantidad de licitaciones por Cluster
inicialElab.loc[:,'labels'] = kmeansElabora.labels_
conteo = inicialElab.groupby(['labels'], as_index = False)[['rbhCode']].count()
conteo
```

Ahora analizamos los centroides para ver si existen diferencias apreciables y explicables entre los distintos clusters

```python
centroidsA = kmeansElabora.cluster_centers_
headers = np.array(['indGarantia','indRenovacion','indCriterio','indExperiencia','RequisitoNacionalidad','RequisitoGeografico','indvalorEnUTM'])
centroids = pd.DataFrame(data=centroidsA,    # 1st column as index
columns=headers)  # 1st row as the column names
pd.options.display.float_format = '{:,.5f}'.format
centroids
```

Finalmente vemos que los clusters se pueden caracterizar de la siguiente manera:

- **Cluster 0 (A)**: licitaciones que tienden a ser positivas en el indicador de Renovación, con algo de riesgo de Garantía, y con un poco de riesgo en los montos estimados cercanos a montos de control
- **Cluster 1 (B)**: licitaciones que tienden a ser positivas en el indicador de Garantía
- **Cluster 2 (C)**: licitaciones que tienden a ser positivas en el indicador de Garantía a la vez que son positivas en el índice de Criterio, con un poco de riesgo de Requisito Geográfico
- **Cluster 3 (D)**: licitaciones que tienden tener un mayor riesgo en montos cercanos a control que el resto de las licitaciones, pero tienden a ser cero o las sin riesgos.
- **Cluster 4 (E)**: licitaciones que tienden a ser positivas en el indicador de Criterio y se agrega un poco de riesgo de Renovación
